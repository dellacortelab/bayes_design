```python
from protein_mpnn_utils import StructureDataset

# By default pandas isn't in the enviroment that the dockerfile setup 
# so run 'pip install pandas' if running this cell doesn't work
import pandas as pd
```

```python
# This is a good preliminary test before running the save train cell below
# This is much smaller so if it doesn't work don't proceed
# Until you get this one running
dataset_valid = StructureDataset('../../pdb_dict_valid.jsonl', truncate=None, max_length=10000)
df_valid = pd.DataFrame(dataset_valid.data)

# First get all the columns with coords in their name 
# and then ~ gets the reverse of that
filtered_valid = df_valid.loc[:, ~df_valid.columns.str.contains('coords')]
filtered_valid.to_csv('pdb_valid_wout_coords.csv')
```

```python
dataset_train = StructureDataset('../../pdb_dict_train.jsonl', truncate=None, max_length=10000)
df_train = pd.DataFrame(dataset_train.data)
filtered_train = df_train.loc[:, ~df_train.columns.str.contains('coords')]
filtered_train.to_csv('pdb_train_wout_coords.csv')
```
